This Data Set contains 21613 houses and 21 attributes describing each house, specifically showcasing sale prices for King County, which includes Seattle. It includes homes sold between May 2014 and May 2015.

```python
import pandas as pd
```

```python
df = pd.read_csv('kc_house_data.csv')
```

```python
df.shape
```

```python
df.info()
```

all 21613 rows contain no null values so our data set is clean

                     HOUSES BUILT BEFORE VS AFTER YR 2000

```python
countAfter = (df['yr_built'] > 2000).sum()
countBefore = countBefore = (df['yr_built'] < 2000).sum()
print(countBefore," ", countAfter)
```

```python
before = df[df['yr_built'] > 2000]['price'].mean()
after = df[df['yr_built'] < 2000]['price'].mean()
print(before , " ", after)
```

1st conclusion: The average price of houses built before year 2000 are approximately 15.80% more expensive than the average price of houses built after year 2000

           BEDROOMS / PRICE

```python
dfbedrooms = df.groupby('bedrooms').agg({'price': ['max', 'min', 'mean', 'count']}).round(1)
dfbedrooms.columns = ['max_price', 'min_price', 'mean_price', 'count']
dfbedrooms.sort_values('mean_price',ascending=False)
```

2nd conclusion: houses having 6 -> 9 bedrooms are occupying the highest mean price values while the mean price of houses having other number of bedrooms tend to be lower

```python
dfbathrooms = df.groupby('bathrooms').agg({'price': ['max', 'min', 'mean', 'count']}).round(1)
dfbathrooms.columns = ['max_price', 'min_price', 'mean_price', 'count']
dfbathrooms.sort_values('mean_price',ascending=False)
```

3rd conclusion: It's noticeable that as the number of bathrooms increase, the price also increases.
the most expensive house has the most amount of bathrooms

               SQFT_LIVING 

```python
corrSqftLiving = df['sqft_living'].corr(df['price'])
corrSqftLiving
```

as of the result of the correlation btw price and sqft_living we can see a strong positive indicator that the price and sqft_living are almost directly proportional to each other

```python
resultSQF = df.groupby('sqft_living').agg({
    'price': ['max', 'mean'],
    'bedrooms': ['max', 'mean'],
    'bathrooms': ['max', 'mean'],
    'floors': ['max', 'mean']
}).sort_values(by=('price', 'mean'), ascending=False).round(1).head(20)
resultSQF
```

4th conclusion: this data set proves our theory on sqft_living and price almost being directly proportional to each other, 
what's also noticable is that as the sqft_living increases, the mean of the number of bathrooms and bedrooms also increases

            GRADE / PRICE

```python
df[['price', 'grade', 'sqft_living', 'sqft_lot']].groupby('grade').mean().round(1).sort_values('price',ascending=False)
```

5th conclusion: as grade increases the price of the houses greatly increases.

               CONDITION / PRICE

```python
df[['price', 'condition', 'sqft_living', 'sqft_lot','bedrooms']].groupby('condition').mean().round(1).sort_values('price',ascending=False)
```

6th conclusion: similarly, as the condition variable increases, the mean of the house's price noticably increases. as well as the mean of the bedrooms.

            WATER FRONT & VIEW

```python
df[['price', 'waterfront', 'view']].groupby('waterfront')['view'].value_counts(normalize=True).mul(100).rename('percentage').reset_index()
```

all houses with waterfront have view

```python
df[['price', 'waterfront']].groupby('waterfront').agg(['max', 'min', 'mean']).reset_index().round(1)
```

7th conclusion: the min value of houses with waterfont is 285k while the min value of houses without waterfront is 75k.
the difference in mean is very noticable, the mean of houses with waterfront are approximately 213% more expensive than the mean of the houses without waterfront

```python
df[['price', 'view']].groupby('view').agg(['max', 'min', 'mean']).reset_index().round(1)
```

8th conclusion: The min value of a house with a view is ~ 150 k while the min value of a house without a view is 75k , that's 100% increase to start off.
as the view variable increases, the mean, min, and max price of the houses increase greatly

                  FINAL CONCLUSION

In summary, our analysis highlights important factors that affect house prices. Houses built before 2000 tend to cost around 15.80% more than those built later. The number of bedrooms, bathrooms, condition, and house grade influence prices directly. Bigger living spaces (sqft_living) often mean more bathrooms and bedrooms, affecting prices. Waterfront properties and houses with nice views are generally pricier. These insights show how various elements combine to shape real estate values.
